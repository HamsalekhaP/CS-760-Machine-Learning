### Split data into train and test instances

```python
split_idx=116

xMenTrain= xMen[:split_idx]
yMenTrain= yMen[:split_idx]
xMenTest= xMen[split_idx+1:]
yMenTest= yMen[split_idx+1:]

train_size=xMenTrain.shape[0]
test_size=xMenTest.shape[0]

XmonTrain= xMon[:split_idx]
yMonTrain= yMon[:split_idx]
xMonTest= xMon[split_idx+1:]
yMonTest= yMon[split_idx+1:]
```

#### Get data statistics

```python
MenMean = np.sum(yMenTrain)/ yMenTrain.shape[0]
MenStd =  math.sqrt(np.sum((yMenTrain - MenMean)**2)/(yMenTrain.shape[0]-1))
MonMean = np.sum(yMonTrain)/ yMonTrain.shape[0]
MonStd = math.sqrt(np.sum((yMonTrain-MonMean)**2)/(yMonTrain.shape[0]-1))

print('Lake Mendota');print('Mean',MenMean);print('Standard Deviation',MenStd)
print('-----')
print('Lake Monona');print('Mean',MonMean);print('Standard Deviation',MonStd)
```

```python
def analytical_solution(X, Y):
    z= inv(np.matmul(X,X.transpose()))
    beta = np.matmul(np.matmul(z,X),Y)
    return(beta)

#  3 * 118
X = np.concatenate((np.ones((1,train_size)), xMenTrain.transpose(), yMonTrain.transpose()), axis=0)
beta = analytical_solution(X , yMenTrain)
print(beta)
```

```python
##Q4
def calculateMSE(X, Y):
    diff = (np.matmul(X.transpose(),beta) - Y)**2
    return np.sum(diff)/diff.shape[0]

Xtest = np.concatenate((np.ones((1,test_size)), xMenTest.transpose(), yMonTest.transpose()), axis=0)
MseError = calculateMSE(Xtest, yMenTest)
print('mean square error', MseError)

def calculateRsquaredError(X,Y):
    cal_mean = np.sum(Y)/Y.shape[0]
    nr =  np.sum((np.matmul(X.transpose(),beta) - Y)**2)
    #print(np.sum((yMenTest- cal_mean)**2))
    TSS = np.sum((yMenTest - cal_mean)**2)
    ratio = (nr/TSS)
    return 1-ratio
#print(Xtest)
RSquareError= calculateRsquaredError(Xtest, yMenTest)
print('R squared error', RSquareError)
```

```python
##Q5

class LinearRegressionWithGD():
    def __init__(self, num_features):
        self.num_features = num_features
        self.beta = np.zeros((num_features, 1),dtype=np.float64)
    
    def forward(self, x):
        return(np.matmul(x.transpose(), self.beta))
    
    def backward(self, x, yhat, y):
        grad_loss_yhat = 2*( yhat - y )
        grad_yhat_beta = x
        grad_loss_beta = np.matmul(grad_yhat_beta,grad_loss_yhat) / y.shape[0]
        return (-1)*grad_loss_beta
    
def loss(yhat, y):
    return np.mean((yhat - y)**2)

def train(model, x, y, num_epochs, learning_rate=0.1):
    cost = []
    for e in range(num_epochs):
        #### Compute outputs ####
        yhat = model.forward(x)
        #### Compute gradients ####
        negative_grad_beta = model.backward(x, yhat, y)
        #### Update weights ####
        model.beta += learning_rate * negative_grad_beta
        #### Logging ####
        yhat = model.forward(x)
        #print('yat after update',yhat)
        curr_loss = loss(yhat, y)
        print('Epoch: %03d' % (e+1), end="")
        print(' | Objective: %.5f' % curr_loss)
        cost.append(curr_loss)

    return cost
```

### Experiments with epochs and learning rate

```python
model1 = LinearRegressionWithGD(num_features=X.shape[0])
cost = train(model1, 
             X, yMenTrain, 
             num_epochs=10, 
             learning_rate=0.1)
```

### Normalise dataset

```python
#Q7
xMenTrainMean = np.mean(xMenTrain)
xMenTrainStd = np.std(xMenTrain)
xMenTrainNorm = (xMenTrain - xMenTrainMean)/xMenTrainStd

yMonTrainMean = np.mean(yMonTrain)
yMonTrainStd = np.std(yMonTrain)
yMonTrainNorm = (yMonTrain - yMonTrainMean)/yMonTrainStd

X_norm= np.concatenate((np.ones((1,train_size)), xMenTrainNorm.transpose(), yMonTrainNorm.transpose()), axis=0)

modelNormalised1 = LinearRegressionWithGD(num_features=X_norm.shape[0])
cost = train(modelNormalised1, 
             X_norm, yMenTrain, 
             num_epochs=2000, 
             learning_rate=0.1)
```
