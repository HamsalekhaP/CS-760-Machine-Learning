```python
'''obs0=current state
action= next state
obs1= next state
action: choosen action by random or best Q
'''
def Q_learning(table, obs0, obs1, reward_table, action):
    # Take the best value reachable from the state obs1
    best_value, _ = best_action_value(Q_table, obs1)

    # Calculate Q-target value 
    Q_target = reward_table[obs0, action] + GAMMA * best_value

    # Update Q(obs0,action)
    q_val=(1-alpha_size_param) * Q_table[obs0,action] + alpha_size_param * Q_target
    print('new Q value at',obs0,action,'is:',q_val)
    Q_table[obs0,action] = q_val
    #print('updated Q_table',Q_table)
```
